```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csvpath = 'Resources/2016-FCC-New-Coders-Survey-Data.csv'
bootcampDF = pd.read_csv(csvpath, low_memory=False)
```

```python
# Read with pandas

bootcampDF.tail(15)
```

```python
# Inspect all columns
bootcampDF.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Use iloc to accomplish this (remember that you need to take the position of the rows and columns into account)

bcdfColumns = bootcampDF.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
bcdfColumns.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
# Hint use the df.replace function
bcdfColumns["AttendedBootcamp"].replace([0.0,1.0], ['No','Yes'], inplace=True)
bcdfColumns["BootcampFinish"].replace([0.0,1.0], ['No','Yes'], inplace=True)
bcdfColumns["BootcampFullJobAfter"].replace([0.0,1.0], ['No','Yes'], inplace=True)
bcdfColumns["BootcampLoanYesNo"].replace([0.0,1.0], ['No','Yes'], inplace=True)
bcdfColumns["BootcampRecommend"].replace([0.0,1.0], ['No','Yes'], inplace=True)
bcdfColumns["HasChildren"].replace([0.0,1.0], ['No','Yes'], inplace=True)


bcdfColumns.head()
```

```python
# Calculate total number of respondents in survey
# consider using the len() function

totalSurveyed = len(bcdfColumns)
totalSurveyed
```

```python
# Extract rows corresponding only to people who attended a bootcamp
# Filter using df.loc on the AttendedBootcamp column
bcattendees = bcdfColumns.loc[bcdfColumns["AttendedBootcamp"] == "Yes"]
bcattendees.head()
```

```python
# Calculate average age of attendees
# Pull out the `Age` series and take the average of it
avgAge = bcattendees["Age"].mean()
avgAge
```

```python
# Calculate how many people attended a bootcamp
# Count the values of the AttendedBootcamp column
totalAttendees = bcattendees["AttendedBootcamp"].count()
totalAttendees
```

```python
# Calculate how many attendees hold degrees
# It is easier to do this with `value_counts() count values for the SchoolDegree column

schoolDegreeDF = bcattendees["SchoolDegree"].value_counts().to_frame()
schoolDegreeDF
```

```python
# Count the number of records where the person is a degree holder
# There are several ways to approach this. You can look for people who have degrees
# or for people who don't have degrees depending on the value of the SchoolDegree column

withdegree = bcattendees.loc[(bcattendees["SchoolDegree"] == "bachelor's degree") |
                             (bcattendees["SchoolDegree"] == "master's degree (non-professional)") |
                             (bcattendees["SchoolDegree"] == "professional degree (MBA, MD, JD, etc.)") |
                             (bcattendees["SchoolDegree"] == "Ph.D.") |
                             (bcattendees["SchoolDegree"] == "associate's degree"),]
percentagewithdegree = withdegree["SchoolDegree"].value_counts().sum() / totalAttendees
percentagewithdegree
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
attendeegenderDF = bcattendees["Gender"].value_counts().to_frame()
# attendeegenderDF["% of Respondents"] = attendeegenderDF["Gender"] / len(bcattendees)
attendeegenderDF["% of Respondents"] = attendeegenderDF["Gender"] / attendeegenderDF["Gender"].sum()
attendeegenderDF
```

```python
# Calculate percentage of respondents who attended a bootcamp
bcpercentageDF = bcdfColumns["AttendedBootcamp"].value_counts().to_frame()
bcpercentageDF["% of Respondents"] = bcpercentageDF["AttendedBootcamp"] / len(bcdfColumns["AttendedBootcamp"])
bcpercentageDF
```

```python
# Calculate percentage of respondents belonging to each gender
totalgenderDF = bcdfColumns["Gender"].value_counts().to_frame()
totalgenderDF["% of Respondents"] = totalgenderDF["Gender"] / totalgenderDF["Gender"].sum()
totalgenderDF
```

```python
# Calculate percentage of attendees with a college degree
percentageDF = bcattendees.loc[(bcattendees["SchoolDegree"] == "bachelor's degree") |
                               (bcattendees["SchoolDegree"] == "master's degree (non-professional)") |
                               (bcattendees["SchoolDegree"] == "professional degree (MBA, MD, JD, etc.)") |
                               (bcattendees["SchoolDegree"] == "Ph.D.") |
                               (bcattendees["SchoolDegree"] == "associate's degree"),]
percentageDF = percentageDF["SchoolDegree"].value_counts().to_frame()
percentageDF["% of Respondents"] = percentageDF["SchoolDegree"] / len(bcattendees)
percentageDF["% of Respondents"].sum()
```

```python
# Calculate average post-bootcamp salary
postBCsalary = bcdfColumns["BootcampPostSalary"].dropna().mean()
postBCsalary
```

```python
# Create a new table consolidating above calculations
consolidatedDF = pd.DataFrame({"Total Surveyed": [totalSurveyed],
                               "Total Bootcamp Attendees": [totalAttendees],
                               "% Attended Bootcamp":[bcpercentageDF.iloc[1,1]],
                               "Avg. Age": [avgAge],
                               "Has a Degree":percentagewithdegree,
                               "% Male":[attendeegenderDF.iloc[0,1]],
                               "% Female":[attendeegenderDF.iloc[1,1]],
                               "% Non Gender Specific":[(attendeegenderDF.iloc[2,1] + attendeegenderDF.iloc[3,1] + attendeegenderDF.iloc[4,1])],
                               "Post Bootcamp Salary":[postBCsalary]
                              })

consolidatedDF = consolidatedDF.astype("float")
consolidatedDF
```

```python
# Improve formatting before outputting spreadsheet
consolidatedDF["% Attended Bootcamp"] = consolidatedDF["% Attended Bootcamp"].map("{:.2%}".format)
consolidatedDF["Avg. Age"] = consolidatedDF["Avg. Age"].map("{:.0f}".format)
consolidatedDF["Has a Degree"] = consolidatedDF["Has a Degree"].map("{:.2%}".format)
consolidatedDF["% Male"] = consolidatedDF["% Male"].map("{:.2%}".format)
consolidatedDF["% Female"] = consolidatedDF["% Female"].map("{:.2%}".format)
consolidatedDF["% Non Gender Specific"] = consolidatedDF["% Non Gender Specific"].map("{:.2%}".format)
consolidatedDF["Post Bootcamp Salary"] = consolidatedDF["Post Bootcamp Salary"].map("${:.2f}".format)
consolidatedDF
```

```python
# Export to Excel
# Use df.to_excel to export to excel. Don't include the indexes
consolidatedDF.to_excel("Homework2.xlsx", index=False)

```
